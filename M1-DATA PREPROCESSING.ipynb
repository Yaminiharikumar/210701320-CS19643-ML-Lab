## DATA PREPROCESSING AND DATA CLEANING

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('CTG.csv')
del df['FileName']
del df['Date']
del df['SegFile']
del df['b']
del df['e']
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

var = ['CLASS','NSP']
  
for i in var:
    df[i] = le.fit_transform(df[i]).astype(int)
```
